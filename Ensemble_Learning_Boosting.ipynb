```python
quantile33 = data['price'].quantile(0.33)
quantile66 = data['price'].quantile(0.66)
quantile100 = data['price'].quantile(1)

data['price_cat'] = pd.cut(data['price'],bins=[0,quantile33,quantile66,quantile100],labels=['Low','Medium','High'])
data.drop('price',axis=1,inplace=True)
print(f"data['price_cat'].value_counts :: {data['price_cat'].value_counts()}")
```

```python
data.head()
```

## Model Building

```python
x = data.drop('price_cat',axis=1)
y = data['price_cat']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

print(f"x_train.shape ::: {x_train.shape}")
print(f"x_test.shape ::: {x_test.shape}")
print(f"y_train.shape ::: {y_train.shape}")
print(f"y_test.shape ::: {y_test.shape}")
```

```python
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.fit_transform(y_test)
```

```python
y_train
```

```python
y_test
```

```python
lr = LogisticRegression(random_state=42,class_weight='balanced')
dt = DecisionTreeClassifier(random_state=42,class_weight='balanced')
rf = RandomForestClassifier(n_estimators=500,random_state=42,class_weight='balanced',n_jobs=-1)
```

```python
gb = GradientBoostingClassifier(n_estimators=500,random_state=42)
xgb = xgb.XGBClassifier(n_estimators=500,random_state=42,num_class=3)
```

```python
clf_list = [lr,dt,rf,gb,xgb]
```

```python
train_scores = []
test_scores = []
clf_names = []
for clf in clf_list:
  clf.fit(x_train,y_train)
  train_scores.append(np.round(a=clf.score(x_train, y_train), decimals=2))
  test_scores.append(np.round(a=clf.score(x_test, y_test), decimals=2))
  clf_names.append(clf.__class__.__name__)
print('Success!')

```

```python
pd.DataFrame({'Classifier':clf_names,'Train Score':train_scores,'Test Score':test_scores})
```

```python
param_grid = {
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': list(range(3, 10))
}
gv = GridSearchCV(estimator=gb,param_grid=param_grid,cv=5,n_jobs=-1)
gv.fit(x_train,y_train)
```

```python
best_model=gv.best_estimator_
best_model
```

```python
best_params=gv.best_params_
best_params
```

```python
best_model.fit(x_train,y_train)
y_pred_train = best_model.predict(x_train)
y_pred_test = best_model.predict(x_test)
```

```python
best_model.score(x_train,y_pred_train)
```
