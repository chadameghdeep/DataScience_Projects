```python
import plotly.express as px
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
iris_df = px.data.iris()
iris_df.shape
```

```python
iris_df.head()
```

```python
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
new_iris_df = iris_df[features]
new_iris_df.head()
```

```python
new_iris_df.shape
```

```python
scaler = MinMaxScaler()
scaled_arr = scaler.fit_transform(X=new_iris_df)
```

```python
scaled_arr
```

```python
pca = PCA(n_components=None)
components = pca.fit_transform(scaled_arr)
```

```python
components
```

```python
pd.DataFrame(components).head()
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_
```

```python
labels = {
    str(i): f"PC {i+1} ({var:.1f}%)"
    for i, var in enumerate(pca.explained_variance_ratio_ * 100)
}

fig = px.scatter_matrix(components,
                        labels=labels,
                        dimensions=range(4),
                        color=iris_df["species"])
fig.update_traces(diagonal_visible=False)
fig.show()
```

## PCA LOADINGS

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
```

```python
cumulative_variance
```

```python
var_df = pd.DataFrame(cumulative_variance,
                      index=[f'PCA-{i}' for i in range(1, 5)],
                      columns=['Explained Variance'])
```

```python
var_df
```
