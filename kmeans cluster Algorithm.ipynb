```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from afinn import Afinn
import plotly.graph_objs as go                                     
```

```python
data = pd.read_csv(filepath_or_buffer='https://raw.githubusercontent.com/insaid2018/PGPDSAI/main/06%20Term%206%20-%20Machine%20Learning%20Advanced/02%20Module%202/data/ecommerce_reviews.csv')
print(f"Shape of the data:{data.shape}")
data.head()
```

```python
data.dropna(subset = ['Title','Review Text'],inplace=True)
```

```python
print(f"Shape of the data:{data.shape}")
data.head()
```

```python
data.info()
```

```python
affin = Afinn()
data['Review Text'] = data['Review Text'].str.lower()
data['Sentiment Score'] = data.apply(lambda row: affin.score(row['Review Text']),axis=1)
```

## Exploratory Data Analysis

```python
data.groupby(by='Rating').mean()
```

```python
data.corr()
```

```python
scaled_frame = data[['Age','Sentiment Score']]
```

```python
inertia_vals = []
k_vals = [x for x in range(1,16)]
for i in k_vals:
  k_model = KMeans(n_clusters = i,max_iter=500,random_state=42)
  k_model.fit(scaled_frame)
  inertia_vals.append(k_model.inertia_)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = k_vals, y = inertia_vals, mode = 'lines+markers'))
fig.show()
```

```python
k_means = KMeans(n_clusters = 6,max_iter=500,random_state=42)
k_means.fit(scaled_frame)
scaled_frame['Labels'] = k_means.labels_ 
centers = k_means.cluster_centers_
```
